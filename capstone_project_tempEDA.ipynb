```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import numpy as np
```

```python
file_path = "/Users/apple/Downloads/capstoneProject/Crime_Data.csv"
df = pd.read_csv(file_path)
df
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()
missing_values
```

```python
# Evaluate the percentage of missing values for each column
total_records = len(df)
missing_percentage = (missing_values / total_records) * 100
missing_percentage
```

```python
# Display missing values and their percentage for each column
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
print("Missing Values:\n", missing_data)
```

```python
# Remove columns with more than 20% missing values, excluding 'Weapon Desc'
columns_to_drop = missing_data[(missing_data['Percentage'] > 20) & (missing_data.index != 'Weapon Desc')].index
df_cleaned = df.drop(columns=columns_to_drop)
df_cleaned
```

```python
# Identify categorical and numerical columns
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns
numerical_columns = df_cleaned.select_dtypes(include=['int64', 'float64']).columns

# Print categorical and numerical columns
print("Categorical Columns:")
print(categorical_columns)

print("\nNumerical Columns:")
print(numerical_columns)
```

```python
# Fill missing values for remaining columns
for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':
        # For categorical columns, fills missing values with the mode
        mode_value = df_cleaned[column].mode()[0]
        df_cleaned[column].fillna(mode_value, inplace=True)
    else:
        # For numerical columns, fills missing values with the mean
        mean_value = df_cleaned[column].mean() 
        df_cleaned[column].fillna(mean_value, inplace=True)
```

```python
df_cleaned
```

```python
# Check for and remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Convert columns to appropriate data types
df_cleaned['Date Rptd'] = pd.to_datetime(df_cleaned['Date Rptd'])
df_cleaned['DATE OCC'] = pd.to_datetime(df_cleaned['DATE OCC'])
```

```python
# Handling outliers
# Outliers in 'TIME OCC' column
time_mean = df_cleaned['TIME OCC'].mean()
time_std = df_cleaned['TIME OCC'].std()
time_upper_limit = time_mean + 3 * time_std
df_cleaned.loc[df_cleaned['TIME OCC'] > time_upper_limit, 'TIME OCC'] = time_upper_limit
```

```python
# Outliers in 'Vict Age' column
age_mean = df_cleaned['Vict Age'].mean()
age_std = df_cleaned['Vict Age'].std()
age_upper_limit = age_mean + 3 * age_std
df_cleaned.loc[df_cleaned['Vict Age'] > age_upper_limit, 'Vict Age'] = age_upper_limit
```

```python
# Handling outliers in 'LAT' column
# Assuming latitude values should fall within the range of -90 to 90 degrees
lat_lower_limit = -90
lat_upper_limit = 90
df_cleaned.loc[df_cleaned['LAT'] < lat_lower_limit, 'LAT'] = lat_lower_limit
df_cleaned.loc[df_cleaned['LAT'] > lat_upper_limit, 'LAT'] = lat_upper_limit
```

```python
# Handling outliers in 'LON' column
# Assuming longitude values should fall within the range of -180 to 180 degrees
lon_lower_limit = -180
lon_upper_limit = 180
df_cleaned.loc[df_cleaned['LON'] < lon_lower_limit, 'LON'] = lon_lower_limit
df_cleaned.loc[df_cleaned['LON'] > lon_upper_limit, 'LON'] = lon_upper_limit
```

```python
# Removing irrelevant columns
columns_to_drop = ['DR_NO', 'Rpt Dist No', 'Part 1-2', 'Mocodes', 'LOCATION']
df_cleaned = df_cleaned.drop(columns=columns_to_drop, axis=1)
```
